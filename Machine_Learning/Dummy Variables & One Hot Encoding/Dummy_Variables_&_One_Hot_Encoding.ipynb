```python
import pandas as pd
```

```python
file = "https://raw.githubusercontent.com/codebasics/py/master/ML/5_one_hot_encoding/Exercise/carprices.csv"
df = pd.read_csv(file)
df
```

```python
dummies = pd.get_dummies(df["Car Model"])
dummies
```

```python
merged = pd.concat([df,dummies],axis="columns")
merged
```

```python
final = merged.drop(["Car Model","Mercedez Benz C class"],axis="columns")
final
```

```python
x = final.drop(["Sell Price($)"],axis="columns")
x
```

```python
y = final["Sell Price($)"]
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.predict([[69000,6,0,1]])
```

```python
model.predict([[83000,7,0,0]])
```

```python
model.predict([[86000,7,0,1]])
```

```python
model.score(x,y)
```
