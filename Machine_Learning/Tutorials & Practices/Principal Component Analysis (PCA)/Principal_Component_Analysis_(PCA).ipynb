```python
import pandas as pd
url = "https://raw.githubusercontent.com/codebasics/py/master/ML/18_PCA/Exercise/heart.csv"
df = pd.read_csv(url)
df.head(5)
```

```python
df.shape
```

```python
df.describe()
```

##Remove outliers using Z score.

```python
df[df.Cholesterol>(df.Cholesterol.mean()+3*df.Cholesterol.std())]
```

```python
df1 = df[df.Cholesterol<=(df.Cholesterol.mean()+(3*df.Cholesterol.std()))]
df1.shape
```

```python
df1[df1.RestingBP>(df1.RestingBP.mean()+(3*df1.RestingBP.std()))]
```

```python
df2 = df1[df.RestingBP<=(df1.RestingBP.mean()+(3*df1.RestingBP.std()))]
```

```python
df2.shape
```

```python
df2[df2.MaxHR>(df2.MaxHR.mean()+(3*df2.MaxHR.std()))]
```

```python
df2[df2.Oldpeak>(df2.Oldpeak.mean()+(3*df2.Oldpeak.std()))]
```

```python
df3 = df2[df2.Oldpeak<=(df2.Oldpeak.mean()+(3*df2.Oldpeak.std()))]
```

```python
df3.shape
```

##Convert text columns to numbers using label encoding and one hot encoding

```python
df3.Sex.unique()
```

```python
df3.ChestPainType.unique()
```

```python
df3.RestingECG.unique()
```

```python
df3.ExerciseAngina.unique()
```

```python
df3.ST_Slope.unique()
```

```python
df4 = df3.copy()
df4.ChestPainType.replace(
    {
        'ATA':1,
        'NAP':2,
        'ASY':3,
        'TA':4
    },
    inplace=True
)
df4.RestingECG.replace(
    {
        'Normal':1,
        'ST':2,
        'LVH':3
    },
    inplace=True
)
df4.ExerciseAngina.replace(
    {
        'N':0,
        'Y':1
    },
    inplace=True
)
df4.ST_Slope.replace(
    {
        'Up':1,
        'Flat':2,
        'Down':3
    },
    inplace=True
)
```

```python
df4.head(5)
```

```python
df5 = pd.get_dummies(df4,drop_first=True)
df5.head(5)
```

```python
X = df5.drop('HeartDisease',axis='columns')
y = df5.HeartDisease
```

```python
X.head()
```

##Apply scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2,random_state=30)
```

###Build a classification model using various methods (SVM, logistic regression, random forest) and check which model gives you the best accuracy

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
m_svm = svm.SVC()
m_svm.fit(X_train,y_train)
m_svm.score(X_test,y_test)
```

```python
m_rfc = RandomForestClassifier()
m_rfc.fit(X_train,y_train)
m_rfc.score(X_test,y_test)
```

```python
m_lr = LogisticRegression()
m_lr.fit(X_train,y_train)
m_lr.score(X_test,y_test)
```

##Use PCA to reduce dimensions

```python
X
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca,y,test_size=0.2,random_state=30)
```

```python
m_rfc = RandomForestClassifier()
m_rfc.fit(X_train_pca,y_train)
m_rfc.score(X_test_pca,y_test)
```
