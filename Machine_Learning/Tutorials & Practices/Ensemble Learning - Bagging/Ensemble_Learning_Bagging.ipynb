```python
import pandas as pd
url = 'https://raw.githubusercontent.com/codebasics/py/master/ML/19_Bagging/Exercise/heart.csv'
df = pd.read_csv(url)
df.head()
```

##Remove outliers using Z score

```python
df[df.RestingBP>(df.RestingBP.mean()+(3*df.RestingBP.std()))]
```

```python
df1 = df[df.RestingBP<=(df.RestingBP.mean()+(3*df.RestingBP.std()))]
df1.shape
```

```python
df1[df1.Cholesterol	>(df1.Cholesterol	.mean()+(3*df1.Cholesterol.std()))]
```

```python
df2 = df1[df1.Cholesterol	<=(df1.Cholesterol	.mean()+(3*df1.Cholesterol.std()))]
df2.shape
```

```python
df2[df2.MaxHR	>(df2.MaxHR	.mean()+(3*df2.MaxHR.std()))]
```

```python
df2[df2.Oldpeak	>(df2.Oldpeak	.mean()+(3*df2.Oldpeak.std()))]
```

```python
df3 = df2[df2.Oldpeak	<=(df2.Oldpeak	.mean()+(3*df2.Oldpeak.std()))]
df3.shape
```

```python
df3.describe()
```

```python
df3.head()
```

##Convert text columns to numbers using label encoding and one hot encoding

```python
df3.ChestPainType.unique(),df3.RestingECG.unique(),df3.ExerciseAngina.unique(),df3.ST_Slope.unique()
```

```python
df4 = df3.copy()
df4.ChestPainType.replace(
    {
        'ATA':0,
        'NAP':1,
        'ASY':2,
        'TA':3
    },
    inplace=True
)
df4.RestingECG.replace(
    {
        'Normal':0,
        'ST':1,
        'LVH':2
    },
    inplace=True
)
df4.ExerciseAngina.replace(
    {
        'N':0,
        'Y':1
    },
    inplace=True
)
df4.ST_Slope.replace(
    {
        'Up':0,
        'Flat':1,
        'Down':2
    },
    inplace=True
)
```

```python
df4.head()
```

```python
df4 = pd.get_dummies(df4,drop_first=True)
df4.head()
```

```python
X = df4.drop('HeartDisease',axis='columns')
y = df4.HeartDisease
X.head()
```

##Apply scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled[0]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.2)
```

####Build a classification model using support vector machine. Use standalone model as well as Bagging model and check if you see any difference in the performance.

```python
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
from sklearn.ensemble import BaggingClassifier
```

```python
scores = cross_val_score(SVC(), X, y, cv=5)
scores 
```

```python
scores.mean()
```

```python
m_bagging = BaggingClassifier(
    base_estimator=SVC(),
    n_estimators=100,
    max_samples=0.8,
    oob_score=True,
    random_state=0
)
m_bagging.fit(X_train,y_train)
```

```python
m_bagging.oob_score_
```

```python
m_bagging.score(X_test,y_test)
```

#####Use decision tree classifier. Use standalone model as well as Bagging and check if you notice any difference in performance

```python
from dataclasses import make_dataclass
from sklearn.tree import DecisionTreeClassifier
scores = cross_val_score(DecisionTreeClassifier(),X,y,cv=10)
scores
```

```python
scores.mean()
```

```python
m_bagging = BaggingClassifier(
    base_estimator=DecisionTreeClassifier(),
    n_estimators=100,
    max_samples=0.8,
    oob_score=True,
    random_state=0
)
m_bagging.fit(X_train,y_train)
m_bagging.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
m_rfc = RandomForestClassifier()
m_rfc.fit(X_train,y_train)
m_rfc.score(X_test,y_test)
```
