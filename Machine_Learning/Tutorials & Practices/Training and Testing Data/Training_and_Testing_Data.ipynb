```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
file = 'https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/Exercise/canada_per_capita_income.csv'
df = pd.read_csv(file)
```

```python
df.head(5)
```

```python
x=df["year"]
x
```

```python
y = df["per capita income (US$)"]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(df[["year"]],df[["per capita income (US$)"]],test_size=0.2,random_state=10)
```

```python
y_train
```

```python
reg = linear_model.LinearRegression()
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```
